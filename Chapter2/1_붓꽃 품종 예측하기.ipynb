```python
# iris 데이터
from sklearn.datasets import load_iris
# pandas
import pandas as pd
```

```python
# 데이터 불러오기
df_raw = load_iris()
df_raw
```

```python
# Target
df_y = df_raw.target
# Features
df_x = df_raw.data
# pandas 변환
df = pd.DataFrame(data = df_x, columns = df_raw.feature_names)
df['Target'] = df_y
df.head()
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
# 데이터 분할
X_train, X_test, Y_train, Y_test = train_test_split(df_x, df_y, test_size = 0.2, random_state = 11)
# 결과 확인
print('train data X size :', X_train.shape)
print('train data Y size :', Y_train.shape)
print('test data X size :', X_test.shape)
print('test data Y size :', Y_test.shape)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier as DTC
# 모델 생성
dtc = DTC(random_state = 11)
# 학습
dtc.fit(X_train, Y_train)
# 예측
y_pred = dtc.predict(X_test)
```

```python
# 성능 평가
from sklearn.metrics import accuracy_score
# 성능 확인
print('정확도 : {:.4f}'.format(accuracy_score(Y_test, y_pred)))
```
